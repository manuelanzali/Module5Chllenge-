### Analysis

- After measuring the development of SCC tumors over 45 days, the Capomulin treatment regimen produced the lowest final tumor volumes across four of the treatment regimens: Capomulin, Ramicane, Infubinol, and Ceftamin; which showed its success as the drug of interest against SCC tumors.
- There was also a positive correlation between the weight of each mouse treated with Capomulin in the study and their average tumor volume, shown by the calculated correlation of 0.84.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

study_results['index_col'] = study_results.index

# Combine the data into a single DataFrame
combined_study_df = pd.merge(study_results, mouse_metadata, how="outer", on="Mouse ID")

combined_study_df.sort_values('index_col', inplace=True)
combined_study_df = combined_study_df.drop(columns=['index_col'])

# Display the data table for preview
combined_study_df = combined_study_df[["Mouse ID","Timepoint","Tumor Volume (mm3)", "Metastatic Sites", 
                                       "Drug Regimen", "Sex", "Age_months", "Weight (g)"]]

combined_study_df.head()
```

```python
# Checking the number of mice.
number_mice = combined_study_df["Mouse ID"].value_counts()
len(number_mice)
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.

duplicates = combined_study_df[combined_study_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]

#duplicate_mice = duplicates['Mouse ID'].nunique()
duplicate_mouse_id = duplicates['Mouse ID'].unique()
duplicate_mouse_id
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_df = combined_study_df.loc[combined_study_df["Mouse ID"] == "g989"]
duplicate_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_combined_study = combined_study_df[combined_study_df['Mouse ID'] != "g989"].copy()
clean_combined_study.head()
```

```python
# Checking the number of mice in the clean DataFrame.
number_clean_mice = clean_combined_study["Mouse ID"].value_counts()
len(number_clean_mice)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
stats = clean_combined_study.groupby("Drug Regimen")["Tumor Volume (mm3)"]
mean = stats.mean()
median = stats.median()
variance = stats.var()
standard_deviation = stats.std()
standard_error = stats.sem()

# Assemble the resulting series into a single summary DataFrame.

#create a new empty dataframe
summary_clean_df = pd.DataFrame()

#name each new column in the new dataframe
summary_clean_df["Mean Tumor Volume"] = mean
summary_clean_df["Median Tumor Volume"] = median
summary_clean_df["Tumor Volume Variance"] = variance
summary_clean_df["Tumor Volume Std. Dev."] = standard_deviation
summary_clean_df["Tumor Volume Std. Err."] = standard_error

#assemble new columns in the order needed
summary_clean_df = summary_clean_df[["Mean Tumor Volume", "Median Tumor Volume", "Tumor Volume Variance", 
                                     "Tumor Volume Std. Dev.", "Tumor Volume Std. Err."]]

summary_clean_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
advanced_stats_df = pd.DataFrame(clean_combined_study.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":[
                                                                                        "mean","median", "var","std", "sem"]}))

advanced_stats_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

mouse_final_timepoint = clean_combined_study.groupby('Mouse ID')['Timepoint'].max()
mouse_final_timepoint = mouse_final_timepoint.reset_index()
final = pd.merge(mouse_final_timepoint,clean_combined_study,on=['Mouse ID','Timepoint'],how="left")
final

```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.
for treatment in treatments:

    # add subset
    subset = final[final['Drug Regimen'] == treatment]['Tumor Volume (mm3)']

    # Locate the rows which contain mice on each drug and get the tumor volumes
    quartiles = subset.quantile([0.25,0.5,0.75])
    Q1 = quartiles[0.25]
    Q3 = quartiles[0.75]
    iqr = Q3-Q1

    # Determine outliers using upper and lower bounds
    lower_bound = Q1 - (1.5*iqr)
    upper_bound = Q3 + (1.5*iqr)

    tumor_vol_data.append(treatments)
    

    outliers = subset[(subset < lower_bound) | (subset > upper_bound)]
    if not outliers.empty:
        print(f"{treatment}'s potential outliers:")
        print(outliers)
    else:
        print(f"{treatment}'s potential outliers: {outliers}")

    #print(f"{treatment}'s potential outliers: Lower bound: {lower_bound:.2f}, Upper bound: {upper_bound:.2f}")


    #print(f"Capomulin's potential outliers: {lower_bound} {upper_bound}")
    #print(f"Ramicane's potential outliers: {lower_bound} {upper_bound}")
    #print(f"Infubinol's potential outliers: {lower_bound} {upper_bound}")
    #print(f"Ceftamin's potential outliers: {lower_bound} {upper_bound}")

#Series([], Name: Tumor Volume (mm3), dtype: float64)
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
tumor_vol_data = []
for treatment in treatments:
    subset = final[final['Drug Regimen'] == treatment]['Tumor Volume (mm3)']
    tumor_vol_data.append(subset)
    
plt.figure(figsize=(12,6))
fig, ax = plt.subplots()
box_plot = ax.boxplot(tumor_vol_data, labels=treatments, flierprops=dict(marker='o', markerfacecolor='r', markersize=12))
ax.set_xlabel("Drug Regimen")
ax.set_ylabel("Final Tumor Volume (mm3)")

#set xtick labels
needed_treatments = len(treatments)
ax.set_xticks(range(1, needed_treatments + 1))
ax.set_xticklabels(treatments, rotation=0, ha='center')
plt.tight_layout()

#for treatment in treatments:
    #subset = final[final['Drug Regimen'] == treatment]['Tumor Volume (mm3)']

    #plt.subplots()
    #plt.boxplot(subset)
```
